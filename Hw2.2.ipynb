```python
print("Task #10")

final = pd.read_csv("2017_jun_final.csv")
python_data = final[final['Язык.программирования'].str.contains('Python', na=False)]
print(python_data)
```

```python
print("Task #11")
print(python_data.shape)
```

```python
print("Task #12")

grouped_by_position = python_data.groupby('Должность')
for position, data in grouped_by_position:
    print(f"Position: {position}")
    print(data)
```

```python
print("Task #13")
aggregated_salaries = python_data.groupby('Должность')['Зарплата.в.месяц'].agg(['min', 'max'])
print(aggregated_salaries)
```

```python
print("Task #14")

def fill_avg_salary(row):
    numeric_values = pd.to_numeric(row, errors='coerce')
    return numeric_values.mean()

python_data.loc[:, 'avg'] = python_data.apply(fill_avg_salary, axis=1)

print(python_data)
```

```python
print("Task #15")

python_data_avg = python_data['avg'].describe()
print(python_data_avg)
```
