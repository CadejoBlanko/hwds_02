```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("Task #1")
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"

pou = pd.read_html(url, attrs={"class": "wikitable collapsible collapsed"})
pou = pou[3]
print(pou.head())
```

```python
print("Task #2")
shape_pou = pou.shape
print(f"Кількість рядків: {shape_pou[0]}")
print(f"Кількість стовпців: {shape_pou[1]}")
```

```python
print("Task #3")
pou.replace("—", np.nan, inplace=True)
print(pou)
```

```python
print("Task #4")
print(pou.dtypes)
```

```python
print("Task #5")
int_pou = pou.fillna(0)
print(int_pou)
```

```python
print("Task #6")
sum_pou = pou.isnull().sum()
print(sum_pou)
```

```python
print("Task #7")
drop_pou = int_pou.drop([27])
print(drop_pou)
```

```python
print("Task #8")
for col in pou.columns[1:]:
    pou[col] = pou[col].fillna(pou[col].astype(float).mean())
print(pou)
```

```python
print("Task #9")
mean_birth_rate_2019 = pou['2019'].astype(float).mean()

regions_above_mean_birth_rate = pou[pou['2019'].astype(float) > mean_birth_rate_2019]['Регіон'].tolist()

print(regions_above_mean_birth_rate)
```

```python
print("Task #10")
pou['2014'] = pd.to_numeric(pou['2014'], errors='coerce')
max_birth_rate_2014 = pou['2014'].max()
region_highest_birth_rate_2014 = pou[pou['2014'] == max_birth_rate_2014]['Регіон'].values[0]
print(f"Регіон з високою народжуваністю у 2014 році: {region_highest_birth_rate_2014}")
```

```python
print("Task #11")
pou.set_index('Регіон', inplace=True)

births_2019 = pou['2019'].astype(float)

plt.figure(figsize=(12, 6))
births_2019.sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Кількість народжень')
plt.title('Народжуваність по регіонах у 2019 році')
plt.show()
```
